```python
def get_zip_df(address):
    name=address
    url = f'https://api.foursquare.com/v2/venues/search?client_id={CLIENT_ID}&client_secret={CLIENT_SECRET}&near={name}&oauth_token={ACCESS_TOKEN}&v={VERSION}&radius={radius}&limit={1000}'
    response = requests.get(url)
    if response.ok:
        try:
            results = response.json()['response']['venues']
            df = pd.json_normalize(results)
        except:
            df = pd.DataFrame()
        return df
    else:
        print(response)
        if response.status_code == 403:
            urllib.error.HTTPError as err:
                print(err.code) 
        return pd.DataFrame()


# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
    
def get_venue_dist(df, zipcode):
    df=df[df["location.postalCode"]==zipcode].copy()
    df['venue_type'] = df.apply(get_category_type, axis=1)
    venue_num = df.shape[0]
    venue_count = dict(df['venue_type'].value_counts())
    return (venue_num, venue_count)
```

## Pull data from API (get venues for each zip)

```python
venue_nums = []
venue_dists = []
for i, row in tqdm(data.iterrows()):
    if i>124:
        address = row.address
        zipcode = str(row.zip)
        df=get_zip_df(address)
        if df.shape[0] != 0:
            venue_num, venue_count = get_venue_dist(df, zipcode)
            venue_nums.append(venue_num)
            venue_dists.append(venue_count)
        else:
            print(i)
            venue_nums.append(0)
            venue_dists.append({})
```
